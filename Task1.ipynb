```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
```

```python
import pandas as pd
df = pd.read_csv("world_population_data.csv")
print(df)
```

```python
df.head()
print(df.head())
```

```python
df = df.reset_index()
df = df.dropna()
```

```python
# Define indicator for total population
indicator = {'SP.POP.TOTL': 'Population'}
print(indicator)
```

```python
# Use the most recent complete year
from datetime import datetime
data_year = datetime.today().year - 1
print(data_year)

data_date = (datetime(2024, 1, 1), datetime(2024, 12, 31))
print(data_date)
```

```python
# Get top 10 most populous countries
top10 = df.groupby('country').max().sort_values(by='Population', ascending=False).head(10).reset_index()
print(top10)
```

```python
#Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='Population', data=top10, palette='viridis')
plt.title('Top 10 Most Populous Countries')
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```

```python
#Histogram of All Country Populations
all_data = df.groupby('country').max().reset_index()

plt.figure(figsize=(10, 6))
sns.histplot(all_data['Population'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Country Populations')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
